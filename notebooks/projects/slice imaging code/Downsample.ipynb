#### Define your movie path, load the data and plot an example frame

## If you have very fine spatial resolution, you can downsample to improve caiman timing
---

#### First, lets visualize the effect of dowsampling our data

```python
# downsample
movie_object = pu.modify_movie(movie_path)
movie_object.spatial_downsample(int(2))

# plt examples
idx = 0
plt.subplot(1,2,1)
plt.imshow(movie[idx,:,:])
plt.subplot(1,2,2)
plt.imshow(movie_object.movie[idx,:,:])
```
